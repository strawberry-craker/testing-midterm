```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

```

```python
trv = pd.read_csv("TRV.csv")
cnc = pd.read_csv("CNC.csv")
eqr = pd.read_csv("EQR.csv")
ivz = pd.read_csv("IVZ.csv")

companies = {"trv" : trv, "cnc" : cnc, "eqr" : eqr, "ivz" : ivz}
```

```python
for df in companies.values() :
    df["Date"] = pd.to_datetime(df["Date"])
    print(df.Date.dtypes)
```

```python
#question 1

for firm, df in companies.items() :
    if sum(df["Close"] == df["Adj Close"]) == df.shape[0]:
        print(f"the company {firm} has no difference in the close and adj close")
```

```python
#question 2

for firm, df in companies.items():
    print(f"the highest adj close price for company {firm} is {df['Adj Close'].max()}")
    print(f"the lowest ad close price for company {firm} is {df['Adj Close'].min()}")
```
